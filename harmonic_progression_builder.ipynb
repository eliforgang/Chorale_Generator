```python
class Note:
    def __init__(self, n):
        self.n=n
        
    def print_note(self):
        print(self.n)

    def notonum(self): #note to number
        if self.n=='c' or self.n=='bs':
            return 1
        elif self.n=='cs' or self.n=='dl':
            return 2
        elif self.n=='d':
            return 3
        elif self.n=='ds' or self.n=='el':
            return 4
        elif self.n=='e' or self.n=='fl':
            return 5
        elif self.n=='es' or self.n=='f':
            return 6
        elif self.n=='fs' or self.n=='gl':
            return 7
        elif self.n=='g':
            return 8
        elif self.n=='gs' or self.n=='al':
            return 9
        elif self.n=='a':
            return 10
        elif self.n=='as' or self.n=='bl':
            return 11
        elif self.n=='b' or self.n=='cl':
            return 12
        else:
            return 'please input a note a-g with flat written as [note]l and sharp written as [note]s'

    def interval(self, other): #find interval between
        other=Note(other)
        if isinstance(self.notonum(), int)==False or isinstance(other.notonum(), int)==False:
            return 'please input a note a-g with flat written as [note]l and sharp written as [note]s'
        elif self.notonum()>other.notonum():
            interval=other.notonum()-self.notonum()+12
        else:
            interval=other.notonum()-self.notonum()
        if interval==1:
            return 'm2'
        elif interval==2:
            return 'M2'
        elif interval==3:
            return 'm3'
        elif interval==4:
            return 'M3'
        elif interval==5:
            return 'P4'
        elif interval==6:
            return 'tritone'
        elif interval==7:
            return 'P5'
        elif interval==8:
            return 'm6'
        elif interval==9:
            return 'M6'
        elif interval==10:
            return 'm7'
        elif interval==11:
            return 'M7'
        else:
            return 'octave/same note'
```

```python
def remove_dup(x):#input string, remove duplicate notes, distinguish between sharps/flats/naturals
    y=[]
    i=0
    z=0
    while i<(len(x)-1): #all but the last character, that way can check i+1
        if x[i+1]=='l' or x[i+1]=='s':
            if (x[i]+x[i+1]) not in y:
                y.append(x[i]+x[i+1])
            i+=2
        else:
            if x[i] not in y:
                y.append(x[i])
            else:
                if y[-1]==x[i]:
                    z=True
                else:
                    for j in range(0,len(y)-1): #check if all instances of a-g occur with flat or sharp
                        if y[j]==x[i] and y[j+1]!='s' and y[j+1]!='l':
                            z=True
                            break
                        else:
                            z=False
                if z==False:
                    y.append(x[i])
            i+=1
    if x[-1] not in y and x[-1]!='l' and x[-1]!='s':
        y.append(x[-1])
    return y
```

```python
class Key:

    def __init__(self, key):
        self.k = key

    def rel_min_maj(self): #find the relative minor/major
        just_note=self.k[:-1]
        major=['a','bl','b','c','cs','dl','d','el','e','f','fs','gl','g','al']
        minor=['fs','g','gs','a','as','bl','b','c','cs','d','ds','el','e','f']
        if self.k[-1]=='M':
            index = major.index(just_note)
            return minor[index]
        else:
            index = minor.index(just_note)
            return major[index]

    def scale(self): #define all scales, output notes of the scales
        if self.k[-1]=='m':
            keysig=Key(self.k).rel_min_maj()
        else:
            keysig=self.k[:-1]
        keysig_set=['a','bl','b','c','cs','dl','d','el','e','f','fs','gl','g','al']
        scale_set=[['a','b','cs','d','e','fs','gs'],['bl','c','d','el','f','g','a'],['b','cs','ds','e','fs','gs','as'],['c','d','e','f','g','a','b'],
                   ['cs','ds','es','fs','gs','as','bs'],
                   ['dl','el','f','gl','al','bl','c'],['d','e','fs','g','a','b','cs'],['el','f','g','al','bl','c','d'],['e','fs','gs','a','b','cs','ds'],
                   ['f','g','a','bl','c','d','e'],['fs','gs','as','b','cs','ds','es'],['gl','al','bl','cl','dl','el','f'],['g','a','b','c','d','e','fs'],
                   ['al','bl','c','dl','el','f','g']]
        scale=scale_set[keysig_set.index(keysig)]
        if self.k[-1]=='m':
            scale=scale[5:]+scale[:5]
        return scale

    def in_key(self, note): #one note as input, determine if note is in the key
        if note in Key(self.k).scale():
            return True
        else:
            return False

    def num_to_rom_num(self,x): #x is number 0-6, the base of the chord
        maj_rn_set = ['I','ii','iii','IV','V','vi','viidim']
        min_rn_set = ['i','iidim','III','iv','v','VI','VII']
        if self.k[-1]=='M':
            rom_num = maj_rn_set[x]
        else:
            rom_num = min_rn_set[x]
        return rom_num
    
    def chord_notes(self,rn): #roman numeral as input, notes as output
        scale = Key(self.k).scale()
        if self.k[-1]=='M':
            if rn=='I':
                chord = [scale[0],scale[2],scale[4]]
            elif rn=='ii':
                chord = [scale[1],scale[3],scale[5]]
            elif rn=='iii':
                chord = [scale[2],scale[4],scale[6]]
            elif rn=='IV':
                chord = [scale[3],scale[5],scale[0]]
            elif rn=='V':
                chord = [scale[4],scale[6],scale[1]]
            elif rn=='vi':
                chord = [scale[5],scale[0],scale[2]]
            elif rn=='viidim':
                chord = [scale[6],scale[1],scale[3]]
            elif rn=='I7':
                chord = [scale[0],scale[2],scale[4],scale[6]]
            elif rn=='ii7':
                chord = [scale[1],scale[3],scale[5],scale[0]]
            elif rn=='iii7':
                chord = [scale[2],scale[4],scale[6],scale[1]]
            elif rn=='IV7':
                chord = [scale[3],scale[5],scale[0],scale[2]]
            elif rn=='V7':
                chord = [scale[4],scale[6],scale[1],scale[3]]
            elif rn=='vi7':
                chord = [scale[5],scale[0],scale[2],scale[4]]
            elif rn=='viihdim7':
                chord = [scale[6],scale[1],scale[3],scale[5]]
            elif rn=='viifdim7':
                if scale[5][-1]=='s':
                    top_note = scale[5][0]
                else:
                    top_note = scale[5] + 'l'
                chord = [scale[6],scale[1],scale[3],top_note]
            else:
                chord = 'invalid input'
        elif self.k[-1]=='m':
            if rn=='i':
                chord = [scale[0],scale[2],scale[4]]
            elif rn=='iidim':
                chord = [scale[1],scale[3],scale[5]]
            elif rn=='III':
                chord = [scale[2],scale[4],scale[6]]
            elif rn=='iv':
                chord = [scale[3],scale[5],scale[0]]
            elif rn=='v':
                chord = [scale[4],scale[6],scale[1]]
            elif rn=='VI':
                chord = [scale[5],scale[0],scale[2]]
            elif rn=='VII':
                chord = [scale[6],scale[1],scale[3]]
            elif rn=='i7':
                chord = [scale[0],scale[2],scale[4],scale[6]]
            elif rn=='iihdim7':
                chord = [scale[1],scale[3],scale[5],scale[0]]
            elif rn=='iifdim7':
                if scale[0][-1]=='s':
                    top_note = scale[0][0]
                else:
                    top_note = scale[0] + 'l'
                chord = [scale[1],scale[3],scale[5],top_note]
            elif rn=='III7':
                chord = [scale[2],scale[4],scale[6],scale[1]]
            elif rn=='iv7':
                chord = [scale[3],scale[5],scale[0],scale[2]]
            elif rn=='v7':
                chord = [scale[4],scale[6],scale[1],scale[3]]
            elif rn=='VI7':
                chord = [scale[5],scale[0],scale[2],scale[4]]
            elif rn=='VII7':
                chord = [scale[6],scale[1],scale[3],scale[5]]
            else:
                chord = 'invalid input'
        else:
            chord = 'invalid input'
        return chord

    def rom_num_5(self,x): #string of notes as input, roman numeral as output. just the three notes in any order, however many repeats
        y = sort_chord(x)
        maj_chord_set = [Key(self.k).chord_notes('I'),Key(self.k).chord_notes('ii'),Key(self.k).chord_notes('iii'),Key(self.k).chord_notes('IV'),
                     Key(self.k).chord_notes('V'),Key(self.k).chord_notes('vi'),Key(self.k).chord_notes('viidim')]
        min_chord_set = [Key(self.k).chord_notes('i'),Key(self.k).chord_notes('iidim'),Key(self.k).chord_notes('III'),Key(self.k).chord_notes('iv'),
                     Key(self.k).chord_notes('v'),Key(self.k).chord_notes('VI'),Key(self.k).chord_notes('VII')]
        maj_rn_set = ['I','ii','iii','IV','V','vi','viidim']
        min_rn_set = ['i','iidim','III','iv','v','VI','VII']
        if self.k[-1]=='m':
            if y in min_chord_set:
                rom_num = min_rn_set[min_chord_set.index(y)]
            else:
                rom_num = 'error'
        else:
            if y in maj_chord_set:
                rom_num = maj_rn_set[maj_chord_set.index(y)]
            else:
                rom_num = 'error'
        return rom_num

    def rom_num_7(self,x): #string of notes as input, roman numeral as output. just the four notes in any order, however many repeats
        y = sort_chord(x)
        maj_chord_set = [Key(self.k).chord_notes('I7'),Key(self.k).chord_notes('ii7'),Key(self.k).chord_notes('iii7'),Key(self.k).chord_notes('IV7'),
                     Key(self.k).chord_notes('V7'),Key(self.k).chord_notes('vi7'),Key(self.k).chord_notes('viihdim7'),Key(self.k).chord_notes('viifdim7')]
        min_chord_set = [Key(self.k).chord_notes('i7'),Key(self.k).chord_notes('iihdim7'),Key(self.k).chord_notes('iifdim7'),Key(self.k).chord_notes('III7'),
                     Key(self.k).chord_notes('iv7'),Key(self.k).chord_notes('v7'),Key(self.k).chord_notes('VI7'),Key(self.k).chord_notes('VII7')]
        maj_rn_set = ['I7','ii7','iii7','IV7','V7','vi7','viihdim7','viifdim7']
        min_rn_set = ['i7','iihdim7','iifdim7','III7','iv7','v7','VI7','VII7']
        if self.k[-1]=='m':
            if y in min_chord_set:
                rom_num = min_rn_set[min_chord_set.index(y)]
            else:
                rom_num = 'error'
        else:
            if y in maj_chord_set:
                rom_num = maj_rn_set[maj_chord_set.index(y)]
            else:
                rom_num = 'error'
        return rom_num

    def V_in_minor(self, x): #determine if major V in minor
        y = remove_dup(x)
        if self.k[-1]=='m':
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            if raised_7 in y and Key(self.k).scale()[1] in y and Key(self.k).scale()[4] in y:
                return True
            elif raised_7 in y and Key(self.k).scale()[3] in y and Key(self.k).scale()[4] in y:
                return True
            else:
                return False
        else:
            return False

    def IV_in_minor(self, x): #determine if major IV in minor
        y = remove_dup(x)
        if self.k[-1]=='m':
            if Key(self.k).scale()[-2][-1]=='l':
                raised_6=Key(self.k).scale()[-2][0]
            else:
                raised_6=Key(self.k).scale()[-2] + 's'
            if raised_6 in y and Key(self.k).scale()[0] in y and Key(self.k).scale()[3] in y:
                return True
            elif raised_6 in y and Key(self.k).scale()[2] in y and Key(self.k).scale()[3] in y:
                return True
            else:
                return False
        else:
            return False
            
    def I_in_minor(self, x): #determine if major I in minor
        y = remove_dup(x)
        if self.k[-1]=='m':
            if Key(self.k).scale()[2][-1]=='l':
                raised_3=Key(self.k).scale()[2][0]
            else:
                raised_3=Key(self.k).scale()[2] + 's'
            if raised_3 in y and Key(self.k).scale()[0] in y and Key(self.k).scale()[4] in y:
                return True
            elif raised_3 in y and Key(self.k).scale()[0] in y and Key(self.k).scale()[6] in y:
                return True
            else:
                return False
        else:
            return False

    def viidim_in_minor(self, x): #determine if viidim in minor
        y = remove_dup(x)
        if self.k[-1]=='m':
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            if Key(self.k).scale()[-2][-1]=='l':
                raised_6=Key(self.k).scale()[-2][0]
            else:
                raised_6=Key(self.k).scale()[-2] + 's'
            if raised_7 in y and Key(self.k).scale()[1] in y and Key(self.k).scale()[3] in y:
                return True
            elif raised_7 in y and Key(self.k).scale()[1] in y and Key(self.k).scale()[5] in y:
                return True
            elif raised_7 in y and Key(self.k).scale()[1] in y and raised_6 in y:
                return True
            else:
                return False
        else:
            return False

    def neapolitan(self, x): #determine if neapolitan chord
        y = remove_dup(x)
        if Key(self.k).scale()[1][-1]=='s':
            lowered_2nd = Key(self.k).scale()[1][0]
        else:
            lowered_2nd = Key(self.k).scale()[1] + 'l'
        if Key(self.k).scale()[3][-1]=='s':
            lowered_4th = Key(self.k).scale()[3][0]
        else:
            lowered_4th = Key(self.k).scale()[3] + 'l'
        if Key(self.k).scale()[3][-1]=='s':
            lowered_6th = Key(self.k).scale()[5][0]
        else:
            lowered_6th = Key(self.k).scale()[5] + 'l'
        if Key(self.k).scale()[3][-1]=='s':
            lowered_1st = Key(self.k).scale()[0][0]
        else:
            lowered_1st = Key(self.k).scale()[0] + 'l'
        notes = [lowered_2nd,lowered_4th,lowered_6th,lowered_1st]
        if notes[0]in y and notes[1] in y and notes[2] in y:
            return True
        elif notes[0]in y and notes[1] in y and notes[3] in y:
            return True
        else:
            return False

    def altered_5(self,x): #determine if altered dominant
        y = remove_dup(x)
        if Key(self.k).scale()[4][-1]=='s':
            lowered5 = Key(self.k).scale()[4][0]
            lowered2 = Key(self.k).scale()[1][0]
            raised5 = Key(self.k).scale()[4] + 's'
            raised2 = Key(self.k).scale()[1] + 's'
        elif Key(self.k).scale()[4][-1]=='l':
            lowered5 = Key(self.k).scale()[4]+'l'
            lowered2 = Key(self.k).scale()[1]+'l'
            raised5 = Key(self.k).scale()[4][0]
            raised2 = Key(self.k).scale()[1][0]
        else:
            lowered5 = Key(self.k).scale()[4]+'l'
            lowered2 = Key(self.k).scale()[1]+'l'
            raised5 = Key(self.k).scale()[4] + 's'
            raised2 = Key(self.k).scale()[1] + 's'
        if self.k[-1]=='m':
            lowered7 = Key(self.k).scale()[6]
            if Key(self.k).scale()[6][-1]=='l':
                raised7 = Key(self.k).scale()[6][:-1] + 's'
            else:
                raised7 = Key(self.k).scale()[6] + 's'
        else:
            if Key(self.k).scale()[6][-1]=='l':
                raised7 = Key(self.k).scale()[6][:-1]
                lowered7 = Key(self.k).scale()[6] + 'l'
            elif Key(self.k).scale()[6][-1]=='s':
                raised7 = Key(self.k).scale()[6] + 's'
                lowered7 = Key(self.k).scale()[6][:-1]
            else:
                raised7 = Key(self.k).scale()[6] + 's'
                lowered7 = Key(self.k).scale()[6] + 'l'     
        lowered_notes = [lowered5,lowered7,lowered2]
        raised_notes= [raised5,raised7,raised2]
        if lowered_notes[0] in y and lowered_notes[1] in y and lowered_notes[2] in y:
            return True
        elif raised_notes[0] in y and raised_notes[1] in y and raised_notes[2] in y:
            return True
        else:
            return False
                
    def inversion_5(self, x): #inversion of 135
        y = remove_dup(x)
        if Key(self.k).viidim_in_minor(x)==True:
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            chord = 'viidim'
            notes = [raised_7,Key(self.k).scale()[1],Key(self.k).scale()[3]]
        elif Key(self.k).V_in_minor(x)==True:
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            chord = 'V'
            notes = [Key(self.k).scale()[4],raised_7,Key(self.k).scale()[1]]
        elif Key(self.k).IV_in_minor(x)==True:
            if Key(self.k).scale()[-2][-1]=='l':
                raised_6=Key(self.k).scale()[-2][0]
            else:
                raised_6=Key(self.k).scale()[-2] + 's'
            chord = 'IV'
            notes = [Key(self.k).scale()[3],raised_6,Key(self.k).scale()[0]]
        elif Key(self.k).I_in_minor(x)==True:
            chord = 'I'
            if Key(self.k).scale()[2][-1]=='l':
                raised_3=Key(self.k).scale()[2][0]
            else:
                raised_3=Key(self.k).scale()[2] + 's'
            notes = [Key(self.k).scale()[0],raised_3,Key(self.k).scale()[4]]
        elif Key(self.k).neapolitan(x)==True:
            if Key(self.k).scale()[1][-1]=='s':
                lowered_2nd = Key(self.k).scale()[1][0]
            else:
                lowered_2nd = Key(self.k).scale()[1] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_4th = Key(self.k).scale()[3][0]
            else:
                lowered_4th = Key(self.k).scale()[3] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_6th = Key(self.k).scale()[5][0]
            else:
                lowered_6th = Key(self.k).scale()[5] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_1st = Key(self.k).scale()[0][0]
            else:
                lowered_1st = Key(self.k).scale()[0] + 'l'
            chord = 'N'
            notes = [lowered_2nd,lowered_4th,lowered_6th]
        elif Key(self.k).altered_5(x)==True:
            z = sort_chord(x)
            if Key(self.k).scale()[4][-1]=='s':
                lowered5 = Key(self.k).scale()[4][0]
                lowered2 = Key(self.k).scale()[1][0]
                raised5 = Key(self.k).scale()[4] + 's'
                raised2 = Key(self.k).scale()[1] + 's'
            elif Key(self.k).scale()[4][-1]=='l':
                lowered5 = Key(self.k).scale()[4]+'l'
                lowered2 = Key(self.k).scale()[1]+'l'
                raised5 = Key(self.k).scale()[4][0]
                raised2 = Key(self.k).scale()[1][0]
            else:
                lowered5 = Key(self.k).scale()[4]+'l'
                lowered2 = Key(self.k).scale()[1]+'l'
                raised5 = Key(self.k).scale()[4] + 's'
                raised2 = Key(self.k).scale()[1] + 's'
            if self.k[-1]=='m':
                lowered7 = Key(self.k).scale()[6]
                if Key(self.k).scale()[6][-1]=='l':
                    raised7 = Key(self.k).scale()[6][:-1]
                else:
                    raised7 = Key(self.k).scale()[6] + 's'
            else:
                if Key(self.k).scale()[6][-1]=='l':
                    raised7 = Key(self.k).scale()[6][:-1]
                    lowered7 = Key(self.k).scale()[6] + 'l'
                elif Key(self.k).scale()[6][-1]=='s':
                    raised7 = Key(self.k).scale()[6] + 's'
                    lowered7 = Key(self.k).scale()[6][:-1]
                else:
                    raised7 = Key(self.k).scale()[6] + 's'
                    lowered7 = Key(self.k).scale()[6] + 'l'     
            lowered_notes = [lowered5,lowered7,lowered2]
            raised_notes= [raised5,raised7,raised2]
            if z==lowered_notes:
                chord = 'Vflat'
                notes = lowered_notes
            else:
                chord = 'V+'
                notes = raised_notes
        else:
            chord = Key(self.k).rom_num_5(y)
            notes = Key(self.k).chord_notes(chord)
        if y[0]==notes[0]:
            return chord
        elif y[0]==notes[1]:
            return chord+'6'
        else: 
            return chord+'64'

    def inversion_7(self, x): #inversion of 1357
        y = remove_dup(x)
        if Key(self.k).viidim_in_minor(x)==True:
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            if Key(self.k).scale()[-2][-1]=='l':
                raised_6=Key(self.k).scale()[-2][0]
            else:
                raised_6=Key(self.k).scale()[-2] + 's'
            if raised_7 in y and Key(self.k).scale()[1] in y and raised_6 in y:
                chord = 'viihdim7'
                notes = [raised_7,Key(self.k).scale()[1],Key(self.k).scale()[3],raised_6]
            else:
                chord = 'viifdim7'
                notes = [raised_7,Key(self.k).scale()[1],Key(self.k).scale()[3],Key(self.k).scale()[5]]
        elif Key(self.k).V_in_minor(x)==True:
            if Key(self.k).scale()[-1][-1]=='l':
                raised_7=Key(self.k).scale()[-1][0]
            else:
                raised_7=Key(self.k).scale()[-1] + 's'
            notes = [Key(self.k).scale()[4],raised_7,Key(self.k).scale()[1],Key(self.k).scale()[3]]
            chord = 'V7'
        elif Key(self.k).IV_in_minor(x)==True:
            if Key(self.k).scale()[-2][-1]=='l':
                raised_6=Key(self.k).scale()[-2][0]
            else:
                raised_6=Key(self.k).scale()[-2] + 's'
            notes = [Key(self.k).scale()[3],raised_6,Key(self.k).scale()[0],Key(self.k).scale()[2]]
            chord = 'IV7'
        elif Key(self.k).I_in_minor(x)==True:
            chord = 'I7'
            if Key(self.k).scale()[2][-1]=='l':
                raised_3=Key(self.k).scale()[2][0]
            else:
                raised_3=Key(self.k).scale()[2] + 's'
            notes = [Key(self.k).scale()[0],raised_3,Key(self.k).scale()[4],Key(self.k).scale()[6]]
        elif Key(self.k).neapolitan(x)==True:
            if Key(self.k).scale()[1][-1]=='s':
                lowered_2nd = Key(self.k).scale()[1][0]
            else:
                lowered_2nd = Key(self.k).scale()[1] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_4th = Key(self.k).scale()[3][0]
            else:
                lowered_4th = Key(self.k).scale()[3] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_6th = Key(self.k).scale()[5][0]
            else:
                lowered_6th = Key(self.k).scale()[5] + 'l'
            if Key(self.k).scale()[3][-1]=='s':
                lowered_1st = Key(self.k).scale()[0][0]
            else:
                lowered_1st = Key(self.k).scale()[0] + 'l'
            chord = 'N7'
            notes = [lowered_2nd,lowered_4th,lowered_6th,lowered1st]
        else:
            chord = Key(self.k).rom_num_7(y)
            notes = Key(self.k).chord_notes(chord)
        ch = chord[:-1]
        if y[0]==notes[0]:
            return chord
        elif y[0]==notes[1]:
            return chord[:-1]+'65'
        elif y[0]==notes[2]:
            return chord[:-1]+'43'
        else: 
            return chord[:-1]+'42'

    def inversion(self, x): #inversion regardless of length
        y = remove_dup(x)
        if len(y)==4:
            chord = Key(self.k).inversion_7(x)
        else:
            chord = Key(self.k).inversion_5(x)
        return chord

    def rom_num_in_key(self, x): #find roman numeral of any chord in the key regardless of length, including inversions
        y = remove_dup(x) #length 1-4
        if len(y) == 1:
            base_note = y[0]
            chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note))
        elif len(y) == 2:
            if Note(y[0]).interval(y[1])[-1]=='3':
                base_note = y[0]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note))
            elif Note(y[0]).interval(y[1])[-1]=='5':
                base_note = y[0]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note))
            elif Note(y[0]).interval(y[1])[-1]=='4':
                base_note = y[1]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note)) + '64'
            elif Note(y[0]).interval(y[1])[-1]=='6':
                base_note = y[1]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note)) + '6'
            elif Note(y[0]).interval(y[1])[-1]=='7':
                base_note = y[0]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note)) + '7'
            else: #interval is 2
                base_note = y[1]
                chord = Key(self.k).num_to_rom_num(Key(self.k).scale().index(base_note)) + '42'
        elif len(y) == 3:
            if Key(self.k).rom_num_5(x)!='error':
                chord = Key(self.k).inversion(x)
            elif Key(self.k).neapolitan(x)==True:
                if Key(self.k).scale()[3][-1]=='s':
                    note = Key(self.k).scale()[5][0]
                else:
                    note = Key(self.k).scale()[5] + 'l'
                if note not in y:
                    y.append(note)
                chord = Key(self.k).inversion(y)
            elif Key(self.k).V_in_minor(x)==True:
                if Key(self.k).scale()[1] not in y:
                    y.append(Key(self.k[:-1]+'M').scale()[1])
                chord = Key(self.k[:-1]+'M').inversion(y)
            elif Key(self.k).IV_in_minor(x)==True:
                if Key(self.k).scale()[0] not in y:
                    y.append(Key(self.k).scale()[0])
                chord = Key(self.k).inversion(y)
            elif Key(self.k).I_in_minor(x)==True:
                if Key(self.k[:-1]+'M').scale()[4] not in y:
                    y.append(Key(self.k[:-1]+'M').scale()[4])
                chord = Key(self.k[:-1]+'M').inversion(y)
            elif Key(self.k).viidim_in_minor(x)==True:
                if Key(self.k[:-1]+'M').scale()[3] not in y:
                    y.append(Key(self.k[:-1]+'M').scale()[3])
                chord = Key(self.k[:-1]+'M').inversion(y)
            elif Key(self.k).altered_5(x)==True:
                chord = Key(self.k).inversion(x)
            else:
                interval1 = Note(y[0]).interval(y[1])
                interval2 = Note(y[1]).interval(y[2])
                interval3 = Note(y[0]).interval(y[2])
                intervals = [interval1,interval2,interval3]
                if 'm2' in intervals or 'M2' in intervals:
                    if interval1[-1] == '2':
                        base = y[1]
                    elif interval2[-1] == '2':
                        base = y[2]
                    else:
                        base = y[2]
                elif 'm7' in intervals or 'M7' in intervals:
                    if interval1[-1] == '7':
                        base = y[0]
                    elif interval2[-1] == '7':
                        base = y[1]
                    else:
                        base = y[0]
                else:
                    base = '0'
                if base!='0':
                    baseinterval1 = Note(base).interval(y[0])
                    baseinterval2 = Note(base).interval(y[1])
                    baseinterval3 = Note(base).interval(y[2])
                    baseintervals = [baseinterval1,baseinterval2,baseinterval3]
                    baseindex = Key(self.k).scale().index(base)
                    if 'm3' or 'M3' in baseintervals:
                        if baseindex<3:
                            y.append(Key(self.k).scale()[int(baseindex + 4)])
                        else:
                            baseindex-=7
                            y.append(Key(self.k).scale()[int(baseindex + 4)])
                        chord = Key(self.k).inversion_7(y)
                    elif 'P5' or 'tritone' in baseintervals:
                        if baseindex<5:
                            y.append(Key(self.k).scale()[int(baseindex + 2)])
                        else:
                            baseindex-=7
                            y.append(Key(self.k).scale()[int(baseindex + 2)])
                        chord = Key(self.k).inversion_7(y)
                    else:
                        chord = 'error'
                else:
                    chord = 'error'
        else:
            if Key(self.k).V_in_minor(x)==True:
                chord = Key(self.k).inversion(x)
            elif Key(self.k).IV_in_minor(x)==True:
                chord = Key(self.k).inversion(x)
            elif Key(self.k).I_in_minor(x)==True:
                chord = Key(self.k).inversion(x)
            elif Key(self.k).viidim_in_minor(x)==True:
                chord = Key(self.k).inversion(x)
            elif Key(self.k).neapolitan(x)==True:
                chord = Key(self.k).inversion(x)
            elif Key(self.k).rom_num_7(x)!='error':
                chord = Key(self.k).inversion_7(x)
            else:
                chord = 'error'
        return chord
        
    def rom_num(self, x): #find roman numeral of any chord in key or V, I, viidim in minor, includes inversions
        y = remove_dup(x)
        chord = ''
        if Key(self.k).V_in_minor(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        elif Key(self.k).IV_in_minor(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        elif Key(self.k).I_in_minor(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        elif Key(self.k).viidim_in_minor(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        elif Key(self.k).neapolitan(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        elif Key(self.k).altered_5(x)==True:
            chord = Key(self.k).rom_num_in_key(x)
        else:
            for i in range(len(y)):
                if Key(self.k).in_key(y[i])==False:
                    if self.k[-2]=='s':
                        lowered = self.k[0]
                    else:
                        lowered = self.k[0]+'l'
                    if self.k[-1]=='M' and y[i]==Key(self.k[:-1]+'m').scale()[5]:
                        chord = ''
                    elif self.k[-1]=='m' and y[i]==lowered:
                        chord = ''
                    else:
                        chord = 'secondary/accidental'
            if chord=='':
                chord = Key(self.k).rom_num_in_key(x)
            elif chord=='secondary/accidental':
                pass
            else:
                chord = 'error'
        return chord
```

```python
def secondary_rom_num(key, x1, x2): #key, current chord, chord immediately after. gives roman numeral of secondary chord
    y1 = remove_dup(x1)
    y2 = remove_dup(x2)
    if Key(key).V_in_minor(x2)==True or Key(key).I_in_minor(x2)==True or Key(key).viidim_in_minor(x2)==True: #move key of next chord to major
        key = key[:-1]+'M'
    chord2 = Key(key).rom_num(x2) #find the roman numeral of next chord
    if chord2 == 'secondary/accidental':
        chord = 'error'
    else:
        if chord2[-2:]=='64' or chord2[-2:]=='65' or chord2[-2:]=='43' or chord2[-2:]=='42': #remove inversions
            chord2 = chord2[:-2]
        elif chord2[-1]=='7' or chord2[-1]=='6':
            chord2 = chord2[:-1]
        if chord2=='viihdim' or chord2=='viifdim':
            chord2 = 'viidim'
        chord2base = Key(key).chord_notes(chord2)[0] #find base note of next chord
        y2new = Key(key).chord_notes(chord2) #find notes of next chord without inversion
        if len(y2new)==4: #remove seventh from denominator
            y2new = y2new[:-1]
        chord2 = Key(key).rom_num(y2new)
        index = Key(key).scale().index(chord2base) #find the degree in the scale
        if key[-1]=='M':
            if index==0 or index==3 or index==4:#establish if secondary key is major or minor
                secondary_key = chord2base + 'M'
            else:
                secondary_key = chord2base  + 'm'
        else:
            if index==2 or index==5 or index==6:
                secondary_key = chord2base + 'M'
            else:
                secondary_key = chord2base  + 'm'
        secondary_scale = Key(secondary_key).scale() #establish scale of secondary key
        chord_num = Key(secondary_key).rom_num(y1) #find the roman numeral of the secondary chord in the secondary key
        if chord_num[0]=='I':
            chord_num='error'
        if chord_num=='error': #indicate if it's not in the key
            chord = 'error'
        else:
            chord = chord_num + '/' + chord2 #secondary chord/chord
    return chord
```

```python
def harmonic_progression(key, chords): #inputs key, list of chords, outputs harmonic progression
    harm_prog = []
    for i in range(len(chords)):
        if Key(key).rom_num(chords[i])!='secondary/accidental' and 'err' not in Key(key).rom_num(chords[i]):
            harm_prog.append(Key(key).rom_num(chords[i]))
        elif secondary_rom_num(key, chords[i], chords[i+1])!='error' and 'secondary/accidental' not in secondary_rom_num(key, chords[i], chords[i+1]):
            harm_prog.append(secondary_rom_num(key, chords[i], chords[i+1]))
        else:
            for j in range(2,len(chords)-i):
                next_chord = harm_prog[-1]
                if secondary_rom_num(key,chords[i],chords[i+j])!='error' and 'secondary/accidental' not in secondary_rom_num(key,chords[i],chords[i+j]):
                    for h in range(1,j):
                        if secondary_rom_num(key,chords[i+h],chords[i+j])!='error' and 'secondary/accidental' not in secondary_rom_num(key,chords[i+h],chords[i+j]):
                            if secondary_rom_num(key,chords[i+h],chords[i+j])[-2:]=='/i':
                                next_chord = secondary_rom_num(key,chords[i],chords[i+j])[:-2]
                            else:
                                next_chord = secondary_rom_num(key,chords[i],chords[i+j])
                if harm_prog[-1]!=next_chord:
                    harm_prog.append(next_chord)
                    break
    return harm_prog
```

```python
def notes(key, chord): #assumes major key
    if '/' not in chord:
        if chord[-1]=='6':
            chord = chord[:-1]
            notes = Key(key).chord_notes(chord)
            base = notes[1]
        elif chord[-2:]=='64':
            chord = chord[:-2]
            notes = Key(key).chord_notes(chord)
            base = notes[2]
        elif chord[-2:]=='65':
            chord = chord[:-2] + '7'
            notes = Key(key).chord_notes(chord)
            base = notes[1]
        elif chord[-2:]=='43':
            chord = chord[:-2] + '7'
            notes = Key(key).chord_notes(chord)
            base = notes[2]
        elif chord[-2:]=='42':
            chord = chord[:-2] + '7'
            notes = Key(key).chord_notes(chord)
            base = notes[3]
        else:
            notes = Key(key).chord_notes(chord)
            base = notes[0]
    else:
        index = chord.index('/')
        denom = chord[index+1:]
        chord = chord[:index]
        if denom=='IV' or denom=='V':
            key = Key(key).chord_notes(denom)[0] + 'M'
            if chord[-1]=='6':
                chord = chord[:-1]
                notes = Key(key).chord_notes(chord)
                base = notes[1]
            elif chord[-2:]=='64':
                chord = chord[:-2]
                notes = Key(key).chord_notes(chord)
                base = notes[2]
            elif chord[-2:]=='65':
                chord = chord[:-2] + '7'
                notes = Key(key).chord_notes(chord)
                base = notes[1]
            elif chord[-2:]=='43':
                chord = chord[:-2] + '7'
                notes = Key(key).chord_notes(chord)
                base = notes[2]
            elif chord[-2:]=='42':
                chord = chord[:-2] + '7'
                notes = Key(key).chord_notes(chord)
                base = notes[3]
            else:
                notes = Key(key).chord_notes(chord)
                base = notes[0]
        else:
            key = Key(key).chord_notes(denom)[0] + 'm'
            if chord[0]=='V':
                if chord[1]=='7' or chord[1:3]=='65' or chord[1:3]=='43' or chord[1:3]=='42':
                    notes = Key(key).chord_notes('v7')
                else:
                    notes = Key(key).chord_notes('v')
                if notes[1][-1]=='l':
                    notes[1] = notes[1][0]
                else:
                    notes[1] = notes[1] + 's'
            elif chord[:2]=='IV':
                if chord[2]=='7' or chord[2:4]=='65' or chord[2:4]=='43' or chord[2:4]=='42':
                    notes = Key(key).chord_notes('iv7')
                else:
                    notes = Key(key).chord_notes('iv')
                if notes[1][-1]=='l':
                    notes[1] = notes[1][0]
                else:
                    notes[1] = notes[1] + 's'
            elif chord[:3]=='vii':
                if chord[3]=='h':
                    notes = Key(key).chord_notes('VII7')
                    if notes[0][-1]=='l':
                        notes[0] = notes[0][0]
                    else:
                        notes[0] = notes[0][0] + 's'
                    if notes[3][-1]=='l':
                        notes[3] = notes[3][0]
                    else:
                        notes[3] = notes[3][0] + 's'
                elif chord[3]=='f':
                    notes = Key(key).chord_notes('VII7')
                    if notes[0][-1]=='l':
                        notes[0] = notes[0][0]
                    else:
                        notes[0] = notes[0][0] + 's'
                else:
                    notes = Key(key).chord_notes('VII')
                    if notes[0][-1]=='l':
                        notes[0] = notes[0][0]
                    else:
                        notes[0] = notes[0][0] + 's'
            else:
                notes = Key(key).chord_notes(chord)
            if chord[-1]=='6' or chord[-2:]=='65':
                base = notes[1]
            elif chord[-2:]=='64' or chord[-2:]=='43':
                base = notes[2]
            elif chord[-2:]=='42':
                base = notes[-1]
            else:
                base = notes[0]
    return notes, base
```

```python
def all_notes(key, chords): #input key and list of chords, output notes and base
    prog = []
    for i in range(len(chords)):
        n, b = notes(key, chords[i])
        prog.append(f'{n} ({b})')
    return prog
```
